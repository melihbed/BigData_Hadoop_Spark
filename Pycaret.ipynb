```python
df=df.replace({"Y":1,"N":0,"BS":1,"MS":2,"PhD":3})
```

```python
df
```

```python
y=df.Hired
x=df.iloc[:,df.columns != "Hired"]
```

```python
from sklearn.tree import DecisionTreeClassifier
d=DecisionTreeClassifier()
```

```python
d.fit(x,y)
```

```python
d.predict([[10,1,4,1,0,1]]) #Ise almamizi soyluyor
```

```python
from sklearn.decomposition import PCA
```

```python
x_y=pca.fit_transform(x)
```

```python
x_y
```

```python
pca.explained_variance_ratio_
```
